```python
import torch
import torchaudio
import torchaudio.transforms as T
from typing import Optional
import matplotlib.pyplot as plt

class ProfessionalAudioProcessor:
    """
    معالج صوتي عالي الدقة لاستخراج Log-Mel Spectrogram.

    المسار الكامل للإشارة:
    Raw Waveform → Resample(16kHz) → Framing(25ms, 10ms stride)
    → Hamming Window → FFT(512) → |Magnitude|² → Mel Filterbank(80)
    → Log(dB)

    المعايير المستخدمة:
    - تردد العينة: 16000 هرتز (معيار VoxCeleb/ECAPA-TDNN)
    - عدد مرشحات ميل: 80 (معيار الصناعة)
    - حجم FFT: 512 نقطة (دقة ترددية 31.25Hz)
    - طول النافذة: 25ms (معيار تحليل الكلام)
    - خطوة النافذة: 10ms (تداخل 60%)

    Reference:
        ECAPA-TDNN: https://arxiv.org/abs/2005.07143
        VoxCeleb: https://www.robots.ox.ac.uk/~vgg/data/voxceleb/
    """

    def __init__(
        self,
        sample_rate: int = 16000,       # Hz - تردد العينة المستهدف
        n_fft: int = 512,               # نقاط FFT
        n_mels: int = 80,               # عدد مرشحات ميل
        frame_length_ms: float = 25.0,  # ms - طول النافذة
        frame_step_ms: float = 10.0,    # ms - خطوة النافذة (للتداخل)
        window_type: str = 'hamming',   # نوع النافذة
        pre_emphasis: float = 0.0,      # معامل Pre-emphasis (0.97 شائع لكن 0 أيضاً مقبول)
        top_db: Optional[float] = None  # اقتطاع الصمت (dB)
    ):
        self.sample_rate = sample_rate
        self.n_fft = n_fft
        self.n_mels = n_mels
        self.frame_length_ms = frame_length_ms
        self.frame_step_ms = frame_step_ms
        self.pre_emphasis = pre_emphasis
        self.top_db = top_db

        # حساب عدد العينات في كل إطار
        self.frame_length = int(frame_length_ms * sample_rate / 1000)  # 400 عينة
        self.hop_length = int(frame_step_ms * sample_rate / 1000)      # 160 عينة

        # اختيار النافذة
        windows = {
            'hamming': torch.hamming_window,
            'hann': torch.hann_window,
            'blackman': torch.blackman_window
        }
        self.window_fn = windows.get(window_type, torch.hamming_window)

        # إنشاء محول الميل مرة واحدة فقط (للكفاءة)
        self.mel_transform = T.MelSpectrogram(
            sample_rate=sample_rate,
            n_fft=n_fft,
            hop_length=self.hop_length,
            n_mels=n_mels,
            window_fn=self.window_fn,
            power=2.0  # مربع المقدار (Power Spectrum)
        )

        # محول إلى ديسيبل
        self.to_db = T.AmplitudeToDB(top_db=top_db)

        # تخزين مرشحات الميل للتوثيق
        self._mel_filterbank = None

    def load_audio(self, file_path: str) -> torch.Tensor:
        """تحميل وتوحيد الملف الصوتي."""
        waveform, sr = torchaudio.load(file_path)

        # Mono conversion
        if waveform.shape[0] > 1:
            waveform = torch.mean(waveform, dim=0, keepdim=True)

        # Resampling
        if sr != self.sample_rate:
            waveform = torchaudio.functional.resample(waveform, sr, self.sample_rate)

        # Pre-emphasis (اختياري)
        if self.pre_emphasis > 0:
            waveform = self._apply_pre_emphasis(waveform)

        return waveform

    def _apply_pre_emphasis(self, waveform: torch.Tensor) -> torch.Tensor:
        """تطبيق مرشح pre-emphasis y[t] = x[t] - α·x[t-1]"""
        return torch.cat([
            waveform[:, :1],
            waveform[:, 1:] - self.pre_emphasis * waveform[:, :-1]
        ], dim=1)

    def extract_features(self, file_path: str) -> torch.Tensor:
        """
        استخراج Log-Mel Spectrogram.

        المخرجات:
            features: (1, n_mels, num_frames)
        """
        # تحميل
        waveform = self.load_audio(file_path)

        # Mel Spectrogram (ينفذ Framing + Windowing + FFT + Mel في خطوة واحدة)
        mel_spec = self.mel_transform(waveform)

        # تحويل إلى ديسيبل
        log_mel = self.to_db(mel_spec)

        return log_mel

    def summary(self) -> str:
        """طباعة ملخص عن المعالج."""
        duration_per_frame = self.frame_length / self.sample_rate * 1000
        freq_resolution = self.sample_rate / self.n_fft

        return f"""
        ╔══════════════════════════════════════╗
        ║   Audio Processor Configuration      ║
        ╠══════════════════════════════════════╣
        ║ Sample Rate:    {self.sample_rate:>5} Hz           ║
        ║ FFT Points:     {self.n_fft:>5} pts           ║
        ║ Freq Res:       {freq_resolution:>5.2f} Hz/bin     ║
        ║ Frame Length:   {self.frame_length_ms:>5.1f} ms ({self.frame_length} samples) ║
        ║ Frame Step:     {self.frame_step_ms:>5.1f} ms ({self.hop_length} samples) ║
        ║ Mel Filters:    {self.n_mels:>5}              ║
        ║ Window Type:    {self.window_fn.__name__:>12}   ║
        ║ Pre-emphasis:   {self.pre_emphasis:>5}              ║
        ╚══════════════════════════════════════╝
        """

    def visualize(self, file_path: str, save_path: Optional[str] = None):
        """تصور الموجة الخام والـ Mel Spectrogram للتوثيق."""
        waveform = self.load_audio(file_path)
        features = self.extract_features(file_path)

        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 6))

        # الموجة الخام
        time_axis = torch.arange(waveform.shape[1]) / self.sample_rate
        ax1.plot(time_axis, waveform.squeeze().numpy(), color='#1f77b4', linewidth=0.5)
        ax1.set_title("Raw Waveform - Time Domain", fontsize=14, fontweight='bold')
        ax1.set_xlabel("Time (seconds)")
        ax1.set_ylabel("Amplitude")
        ax1.grid(alpha=0.3)

        # Mel Spectrogram
        im = ax2.imshow(
            features.squeeze().numpy(),
            aspect='auto',
            origin='lower',
            cmap='magma',
            extent=[0, features.shape[2] * self.frame_step_ms / 1000, 0, self.n_mels]
        )
        ax2.set_title("Log-Mel Spectrogram", fontsize=14, fontweight='bold')
        ax2.set_xlabel("Time (seconds)")
        ax2.set_ylabel("Mel Filter Index")
        plt.colorbar(im, ax=ax2, label='dB')

        plt.tight_layout()

        if save_path:
            plt.savefig(save_path, dpi=150, bbox_inches='tight')
            print(f"✅ تم حفظ الرسم في: {save_path}")

        plt.show()


if __name__ == "__main__":
    processor = ProfessionalAudioProcessor()
    print(processor.summary())
```

```python
import torch
import torchaudio

def extract_speaker_embedding(audio_path, classifier, pre_emphasis_coef=0.97):
    """
    استخراج بصمة المتحدث من ملف صوتي مع معالجة فيزيائية متقدمة وتطبيع الإشارة.

    Args:
        audio_path: مسار الملف الصوتي.
        classifier: نموذج ECAPA-TDNN المُحمّل من SpeechBrain.
        pre_emphasis_coef: معامل مرشح التشديد المسبق (الافتراضي 0.97).

    Returns:
        embedding: متجه البصمة النقي بأبعاد (1, 192).
    """
    # 1. تحميل الصوت
    waveform, sr = torchaudio.load(audio_path)

    # 2. تحويل الصوت إلى أحادي (Mono) إذا كان ستيريو
    if waveform.shape[0] > 1:
        waveform = torch.mean(waveform, dim=0, keepdim=True)

    # 3. توحيد تردد العينات إلى 16000 هرتز
    if sr != 16000:
        waveform = torchaudio.functional.resample(waveform, sr, 16000)

    # 4. تطبيع الصوت (Audio Normalization) - الحل الأساسي لفروق المايكروفونات والمسافة
    # نقوم بتقسيم الإشارة على قيمتها القصوى (Peak Normalization) لتصبح سعة الموجة (Amplitude)
    # موحدة دائماً بين -1.0 و +1.0 في كل الملفات، مهما كان الصوت الأصلي خافتاً أو عالياً.
    if torch.max(torch.abs(waveform)) > 0:
        waveform = waveform / torch.max(torch.abs(waveform))

    # 5. تطبيق مرشح التشديد المسبق (Pre-emphasis Filter) السريع رياضياً
    # الهدف: تصفية ضوضاء الخلفية خفيضة التردد (مثل المكيفات) وتضخيم ترددات الحبال الصوتية العالية.
    if pre_emphasis_coef > 0:
        # y[t] = x[t] - α · x[t-1]
        waveform[:, 1:] = waveform[:, 1:] - pre_emphasis_coef * waveform[:, :-1]

    # 6. استخراج البصمة من النموذج
    # بما أن الإشارة الآن منقاة ومطبعة، ستعمل طبقة الـ Attention داخلياً بأعلى كفاءة
    with torch.no_grad():
        embedding = classifier.encode_batch(waveform)

    return embedding.squeeze()
```

```python
# استخراج بصمتين
emb1 = extract_speaker_embedding("/content/record11.wav", classifier)
emb2 = extract_speaker_embedding("/content/record22.wav", classifier)  # نفس الشخص
emb3 = extract_speaker_embedding("/content/record3.wav", classifier)  # شخص آخر

# حساب التشابه (Cosine Similarity)
similarity_same = torch.nn.functional.cosine_similarity(emb1.squeeze(), emb2.squeeze(), dim=0)
similarity_diff = torch.nn.functional.cosine_similarity(emb1.squeeze(), emb3.squeeze(), dim=0)

print(f"نفس المتحدث:     {similarity_same.item():.3f}")   # يتوقع > 0.7
print(f"متحدثين مختلفين: {similarity_diff.item():.3f}")  # يتوقع < 0.3
```
